```python

s = "Python Programming"
print(f"Q1: {s[:6]}")

print(f"Q2: {'DataScience'[::-1]}")


print(f"Q3: {'z' in 'amazon'}")


f = 3.14
i = int(f)
print(f"Q4: {str(i)} (Type: {type(str(i))})")


L = [10, 20, 30]
L.append(40)
L.insert(1, 15)
print(f"Q5: {L}")

t = (1, 2, 3)
# t[1] = 5
print("Q6: Tuples are immutable and cannot be changed after creation.")

d = {"name": "Alice", "age": 25}
d["age"] = 26
d["city"] = "New York"
print(f"Q7: {d}")

L_dup = [1, 2, 2, 3, 4, 4, 5]
unique_L = list(set(L_dup))
print(f"Q8: {unique_L}")

x = 7
print(f"Q9: {'Even' if x % 2 == 0 else 'Odd'}")


print("Q10:", end=" ")
for i in range(1, 16):
    if i % 3 == 0 and i % 5 == 0:
        print("FizzBuzz")
    elif i % 3 == 0:
        print("Fizz")
    elif i % 5 == 0:
        print("Buzz")
    else:
        print(i)


count, total = 1, 0
while count <= 10:
    total += count
    count += 1
print(f"Q11: Total sum is {total}")

# 12. Nested Loops
print("Q12:")
for _ in range(3):
    for _ in range(3):
        print("*", end=" ")
    print()

# 13. Squares Comprehension
squares = [x**2 for x in range(1, 11)]
print(f"Q13: {squares}")

# 14. Filtering Comprehension
evens = [x for x in range(20) if x % 2 == 0]
print(f"Q14: {evens}")

# 15. Conditional Mapping
scores = [45, 88, 72, 30]
results = ["Pass" if s >= 50 else "Fail" for s in scores]
print(f"Q15: {results}")

# 16. Manual Iteration
colors = ["red", "green", "blue"]
it = iter(colors)
print(f"Q16: {next(it)}, {next(it)}")

# 17. Advanced Slicing
nums = list(range(20))
print(f"Q17: {nums[::3]}")

# 18. String Formatting
item, price = "Coffee", 4.9912
print(f"Q18: The price of {item} is ${price:.2f}")

# 19. Dictionary Comprehension
cubes = {x: x**3 for x in range(1, 6)}
print(f"Q19: {cubes}")

# 20. Flattening
nested = [[1, 2], [3, 4], [5, 6]]
flat = [item for sublist in nested for item in sublist]
print(f"Q20: {flat}")

# Q21: mybirds.insert(1, 'Crow')
mybirds = ["Parrot", "Sparrow", "Crow", "Eagle"]
mybirds.insert(1, 'Crow')
print(mybirds)


# Q22: mybirds.remove("Pigeon") — raises ValueError since "Pigeon" is not in the list
mybirds = ["Parrot", "Sparrow", "Crow", "Eagle"]
try:
    mybirds.remove("Pigeon")
except ValueError as e:
    print("Error:", e)
```
