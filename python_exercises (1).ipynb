```python
def q1(s):
    return s[::-1], s[::2]

def q2(nums):
    return [x*x for x in nums if x % 2 == 0]

def q3(s):
    d = {}
    for c in s:
        d[c] = d.get(c, 0) + 1
    return d

def q4(lst):
    return [i for i, x in enumerate(lst) if x % 2 != 0]

def q5(lst, k):
    k = k % len(lst)
    return lst[-k:] + lst[:-k]

def q6(a, b):
    return dict(zip(a, b))

def q7(lst):
    it = iter(lst)
    try:
        while True:
            print(next(it))
    except StopIteration:
        pass

def q8(lst):
    return [x for sub in lst for x in sub]

def q9(s):
    return s == s[::-1]

def q10(lst):
    first = second = float('-inf')
    for x in lst:
        if x > first:
            second = first
            first = x
        elif x > second and x != first:
            second = x
    return second
```
